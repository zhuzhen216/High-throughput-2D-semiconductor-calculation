### Structure of CO_0
Structure of CO before optimization has a 2D rectangular lattice. After optimization, the structure converts to 1D chains. The one-shot HSE calculation has large residue forece on the each atoms, indicating that HSE-optimized structure can have a large difference from that of PBE-optimized one.
<img src="structure_images/CO_0.png">

```python
Semi_2D_info[(Semi_2D_info['HSE-gap (eV)']>1.3) & (Semi_2D_info['Band gap (eV)']<0.5)]
```

## 2D group IV-VI compounds
___

```python
plt.plot(Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Band gap (eV)'],Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['HSE-gap (eV)'],'o')
plt.plot([0,3],[0,3],':')
plt.xlabel('PBE gap (eV)')
plt.ylabel('HSE gap (eV)')
plt.xticks([0,1,2,3])
plt.yticks([0,1,2,3,4])
```

___
### Band gap vs compound mass
___

```python
plt.plot(Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Compound mass'],Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Band gap (eV)'],'bo')
plt.plot(Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Compound mass'],Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['HSE-gap (eV)'],'r^')
plt.xlabel('Compound mass')
plt.ylabel('Band gap (eV)')
```

### Short summary:
1. In general, HSE band gap is larger than PBE band gap;
2. There is no clear relation between Band gap and compound mass for IV-VI materials; Namely, compound mass is not a good descripter.

___
### Band gap vs elec_neg_diff
___

```python
plt.plot(Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Elec_Neg_diff'],Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Band gap (eV)'],'bo')
plt.plot(Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['Elec_Neg_diff'],Semi_2D_info[Semi_2D_info['Compound type']=='IV-VI']['HSE-gap (eV)'],'r^')
plt.xlabel('Electronegativity difference')
plt.ylabel('Band gap (eV)')
```

### Short summary:
1. The relation between band gap and electronegativity difference is not clearly linear.
2. In general, band gap of the material increases with increasing electronegativity difference.
