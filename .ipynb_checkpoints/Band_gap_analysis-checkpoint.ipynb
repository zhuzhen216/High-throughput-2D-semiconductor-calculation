```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import atomic_info
import sklearn
import seaborn as sns
plt.style.use('classic')
%matplotlib inline
#print(plt.style.available)
```

```python
band_gap = open('band_gap.dat','r')
BandGap_cont = band_gap.readlines()
```

```python
BandGap_dic = {}
BandGap_list = []
for line in BandGap_cont:
    BandGap_dic[line.split()[1]]=float(line.split()[2])
    BandGap_list.append((line.split()[0],line.split()[1],float(line.split()[2])))
```

```python
gap_df=pd.DataFrame(BandGap_list,columns=['Compound','Phase','Band gap (eV)'])
#gap_df[gap_df['Compound']=='CSe']
# Sb_2 and Bi_2 needs more!
```

```python
band_gap_info = open('band_gap_info.dat','r')
band_gap_info_cont = band_gap_info.readlines()
VBM_dic = {}
CBM_dic = {}
for line in band_gap_info_cont:
    VBM_dic[line.split()[1]]=float(line.split()[3])
    CBM_dic[line.split()[1]]=float(line.split()[4])
band_gap_info.close()
```

```python
#
# add VBM and CBM into "gap_df"
#
gap_df['VBM (eV)']=gap_df['Phase'].map(lambda x: VBM_dic[x])
gap_df['CBM (eV)']=gap_df['Phase'].map(lambda x: CBM_dic[x])
```

```python
gap_df.head()
```

```python
gap_df.to_csv('2D_semiconductor.csv')
```

```python
#
# vacuum level based on PBE calculations
#
vacuum_level_info = open('VAC_lev.dat','r')
vacuum_level_cont = vacuum_level_info.readlines()
vacuum_level_dic = {}
for line in vacuum_level_cont:
    vacuum_level_dic[line.split()[0]]=float(line.split()[1])
gap_df['Vacuum level (eV)']=gap_df['Phase'].map(lambda x: vacuum_level_dic[x] 
                                                if x in vacuum_level_dic else None)
```

```python
#
# VBM and CBM absolute levels
#
gap_df['VBM absolute (eV)']=gap_df['VBM (eV)']-gap_df['Vacuum level (eV)']
gap_df['CBM absolute (eV)']=gap_df['CBM (eV)']-gap_df['Vacuum level (eV)']
```

```python
#
# vacuum level based on HSE calculations
#
vacuum_level_hse_info = open('VAC_lev_HSE.dat','r')
vacuum_level_hse_cont = vacuum_level_hse_info.readlines()
vacuum_level_hse_dic = {}
for line in vacuum_level_hse_cont:
    vacuum_level_hse_dic[line.split()[0]]=float(line.split()[1])
gap_df['HSE vacuum level (eV)']=gap_df['Phase'].map(lambda x: vacuum_level_hse_dic[x] 
                                                    if x in vacuum_level_hse_dic else None)
```

```python
#
# VBM and CBM absolute levels
#
gap_df['HSE VBM absolute (eV)']=gap_df['HSE-VBM (eV)']-gap_df['HSE vacuum level (eV)']
gap_df['HSE CBM absolute (eV)']=gap_df['HSE-CBM (eV)']-gap_df['HSE vacuum level (eV)']
```

```python
gap_df.head()
```

```python
# HSE results
band_gap_info = open('HSE_band_gap.dat','r')
band_gap_info_cont = band_gap_info.readlines()
HSE_gap_dic = {}
HSE_VBM_dic = {}
HSE_CBM_dic = {}
for line in band_gap_info_cont:
    HSE_gap_dic[line.split()[1]]=float(line.split()[2])
    HSE_VBM_dic[line.split()[1]]=float(line.split()[3])
    HSE_CBM_dic[line.split()[1]]=float(line.split()[4])
band_gap_info.close()
```

```python
gap_df['HSE-gap (eV)']=gap_df['Phase'].map(lambda x: HSE_gap_dic[x])
gap_df['HSE-VBM (eV)']=gap_df['Phase'].map(lambda x: HSE_VBM_dic[x])
gap_df['HSE-CBM (eV)']=gap_df['Phase'].map(lambda x: HSE_CBM_dic[x])
```

```python
#
# calculate the mass of the compound
#
def calc_average_mass(compound_name):
    elem_list=atomic_info.count_elem(compound_name)
    if len(elem_list)==1:
        if elem_list[0]=='Bi':
            return atomic_info.element.atomic_mass['Bis']
        else:
            return atomic_info.element.atomic_mass[elem_list[0]]
    else:
        return 0.5*(atomic_info.element.atomic_mass[elem_list[0]]+
                    atomic_info.element.atomic_mass[elem_list[1]])
#calc_average_mass('GeS')
```

```python
#
# this function can calculate elec_negativity difference in a compound
#
def calc_elec_neg_diff(compound_name):
    elem_list=atomic_info.count_elem(compound_name)
    if len(elem_list)==1:
        return 0
    else:
        return abs(atomic_info.element.atomic_elec_neg[elem_list[0]]-
                   atomic_info.element.atomic_elec_neg[elem_list[1]])
#calc_elec_neg_diff('GeS')
#atomic_info.element.
```

```python
#
# add more features
#
# tag the compound with 'V','IV-VI', and 'IIV-VII'
def obt_group(compound_name):
    elem_list=atomic_info.count_elem(compound_name)
    if len(elem_list)==1:
        return 'V'
    else:
        if compound_name in atomic_info.compound_III_VII:
            return 'III-VII'
        if compound_name in atomic_info.compound_IV_VI:
            return 'IV-VI'
#obt_group('Bis')
```

```python
def obt_radius(compound_name):
    elem_list=atomic_info.count_elem(compound_name)
    if len(elem_list)==1:
        return atomic_info.element.atomic_radius[elem_list[0]]    
    else:
        return 0.5*(atomic_info.element.atomic_radius[elem_list[0]]+
                    atomic_info.element.atomic_radius[elem_list[1]])
#obt_radius('GeS')
```

```python
#
# get the phase of the compound
# take GeS_0 for example:
# input 'GeS_0'
# return 0
#
def obt_phase(phase_string):
    phase = int(phase_string.split('_')[-1])
    return phase
#obt_phase('GeS_0')
```

```python
#
# then, we can use the function above to put electronegativity difference into our dataframe
#
#gap_df['Elec_Neg_diff']=[0]*118
gap_df['Elec_Neg_diff']=gap_df['Compound'].map(lambda x: calc_elec_neg_diff(x))
#gap_df['Elec_Neg_diff']
#gap_df
```

```python
gap_df.head()
```

```python
#
# add compound mass into the data frame
# 
gap_df['Compound mass']=gap_df['Compound'].map(lambda x: calc_average_mass(x))
#gap_df
```

```python
#
# add compound type into the data frame: V, IV-VI, III-VII
#
gap_df['Compound type']=gap_df['Compound'].map(lambda x: obt_group(x))
#gap_df
```

```python
#
# add phase of the compound into the dataframe
#
gap_df['Phase index']=gap_df['Phase'].map(lambda x: obt_phase(x))
#gap_df
```

```python
#
# add average atomic radius of the compound
#
gap_df['Average radius']=gap_df['Compound'].map(lambda x: obt_radius(x))
#gap_df.head()
```

### Linear regression model
___

```python
import sklearn.linear_model as lm
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(gap_df[['Elec_Neg_diff',
                                                            'Compound mass','Phase index',
                                                            'Average radius']],
                                                    gap_df['Band gap (eV)'],test_size=0.2)
gap_linear=lm.LinearRegression()
gap_linear.fit(X_train, Y_train)
predict_gap=gap_linear.predict(X_test)
```

```python
print(gap_linear.coef_)
print(gap_linear.intercept_)
```

```python
plt.plot(Y_test, predict_gap,'o')
plt.ylim((0,3))
plt.xticks([0,1,2,3])
plt.yticks([0,1,2,3])
plt.xlabel('Calculated band gap (eV)')
plt.ylabel('Predicted band gap (eV)')
plt.plot([0,3],[0,3],':')
```

___
### Results analysis for linear regression model:
1. E_gap = 0.36992416ElecNeg_diff -0.00483363Compound_mass -0.22645096Phase_index + 0.02216003Average_Radius - 0.61
2. From above equation, we can still conclude that (a) the band gap would increase along with the electronegativity difference; (b) the heavier the compounds (the larger they will be), the band gaps become smaller. 
3. For the full set of data, linear regression does not work very well for band gap prediction.
